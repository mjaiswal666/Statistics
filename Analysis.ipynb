```python
import pandas as pd
import numpy as np
import random
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=[12,52,63,48,15,25,63,14,25,36,21,45,78,98,65,42,75,54,21,36,32,52]
```

```python
def insertion(df):
    selection=1
    data_range=len(df)
    while selection<data_range:
        for i in range(selection):
            if(df[selection]<df[i]):
                temp=df[i]
                df[i]=df[selection]
                df[selection]=temp
                selection+=1
        return df
        
```
